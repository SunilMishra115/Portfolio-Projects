```python
# FInd correclation between Budget Vs Gross using scatter plot

plt.scatter(x=data['budget'],y=data['gross'])
plt.title('Budget Vs Gross')
plt.xlabel('Budget for flim')
plt.ylabel('Gross Earning')
plt.show() 
```

```python
data.head()
```

```python
sns.regplot(x='budget',y='gross',data=data,scatter_kws={'color':'blue'})
plt.title('budget vs Gross')
```

```python
#correlation between Score Vs Gross
sns.regplot(x='score',y='gross',data=data,scatter_kws={'color':'brown'},line_kws={'color':'blue'})
plt.title('Score Vs Gross')
```

```python
data_correlation=data.corr()
data_correlation
```

```python
#correlation Between all numerical features
sns.heatmap(data_correlation,annot=True)
plt.title('correlation matrix')
plt.xlabel('Movie Feature')
plt.ylabel('Movie Feature')
plt.show()
```

```python
#converting all values in numeric values
movies_numeric = data

for col_name in movies_numeric.columns:
    if(movies_numeric[col_name].dtypes == 'object'):
            movies_numeric[col_name] = movies_numeric[col_name].astype('category')
            movies_numeric[col_name] = movies_numeric[col_name].cat.codes
            
movies_numeric.head(10)  
```

```python
all_data =movies_numeric.corr()
all_data
```

```python
sns.heatmap(all_data,annot=True)
plt.title("Correlation Matrix")
```

```python
correlation_matrix= all_data
correlation_pairs= correlation_matrix.unstack()
correlation_pairs.head(30)
```

```python
correlation_pairs['gross'].sort_values(ascending=False)
```
